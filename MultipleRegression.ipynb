```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.feature_selection import f_regression
from statsmodels.formula.api import ols
import statsmodels.api as sm
from scipy.stats.mstats import zscore
```

```python
cereal = pd.read_csv('cereal_ds.csv')
cereal.head()
target='rating'
reg_df=cereal
exoge_vars=['sugars', 'fat', 'fiber', 'weight', 'vitamins', 'cups']
```

```python
reg_df
```

```python
# Performing Forward Selection to understand the how significant the values are

for exoge_var in exoge_vars:
    model =target+'~' + exoge_var
    res1 = ols(model, data=reg_df).fit()
    print(f'\n  predicting {target} using {exoge_var:5} alone resulting in  R^2  = {res1.rsquared:6.4f} '    )
    print( f' Explained SS {res1.ess}  and total SS={res1.centered_tss}  resid df= {res1.df_resid}  and mse resid of {res1.mse_resid} ')
    print( f' fvalue of {res1.fvalue} and f_pvalue {res1.f_pvalue}  ')
   
```

## Part A

a. The best predicting model for the purpose of 'rating' using only one predictors should be with Sugars 

```python
# only one predictor is SUGARS
res = ols(
    'rating  ~ sugars',
    data=cereal,
).fit()
print(res.summary())
```

## B

Best 2 predictors is with Sugars and Fiber 

```python
# only 2 is SUGARS+FIBER
# Because p value 2.43e, r-sq 0.80 explains good variation in response 
res = ols(
    'rating  ~ sugars+fiber',
    data=cereal,
).fit()
print(res.summary())
```

Part C - Built a Simple regression model, for predicting 'rating', the best combination is which includes sugars, fiber, fat, vitamins, and weight as predictor variables.

```python
# Best model
res = ols(
    'rating  ~ fiber+sugars+fat+weight+vitamins',
    data=cereal,
).fit()
print(res.summary())
```

## D

The forward selection method shows best combinations by adding more predictors to it 


```python
# Now Performing Forward election to demonstrate best combination based on all the values observed. 


from statsmodels.formula.api import ols
exoge_temp_vars=exoge_vars[:]
## R^2 based forward selection
base_model = target+'~'
# using R^2 as selection criteria
model_r2 = []
model_r2_ess=[]
model_r2_ssr=[]
model_r2_df_resid=[]
model_r2_mse=[]
model_r2_vars=[]
model_nvars=[]
models=[]
nvar=0
for cnt in range(1, len(exoge_temp_vars) + 1):
    max_r2 = -1   
    max_r2_ess=-1
    for exoge_temp_var in exoge_temp_vars:
        model = base_model + ' + ' + exoge_temp_var 
        res1 = ols(model, data=reg_df).fit()
        if res1.rsquared > max_r2:
            max_var = exoge_temp_var
            max_r2 = res1.rsquared
            max_r2_ssr=res1.ssr
            max_r2_ess = res1.ess
            max_r2_df_resid=res1.df_resid
            max_r2_mse=res1.mse_resid
    base_model = base_model + ' + ' + max_var
    exoge_temp_vars.remove(max_var)
    model_r2.append(max_r2)
    models.append(base_model)
    model_r2_ess.append(max_r2_ess)
    model_r2_ssr.append(max_r2_ssr)
    model_r2_vars.append(max_var)
    model_r2_df_resid.append(max_r2_df_resid)
    model_r2_mse.append(max_r2_mse)
    model_nvars.append(cnt)
```

```python
from scipy.stats import f
temp_ess=[0]+model_r2_ess
model_sess=[]
for i in range(1,len(temp_ess)):
    model_sess.append(temp_ess[i]-temp_ess[i-1])
print(model_nvars)
forward_sel_df = pd.DataFrame()
forward_sel_df['nvar'] = model_nvars
forward_sel_df['model'] = models
forward_sel_df['add_var'] =model_r2_vars
forward_sel_df['r2'] = model_r2
forward_sel_df['ess'] = model_r2_ess
forward_sel_df['sess'] =model_sess
forward_sel_df['df_resid']=model_r2_df_resid
forward_sel_df['model_r2_mse']=model_r2_mse
par_fvalue=forward_sel_df['sess']/forward_sel_df['model_r2_mse']
par_f_pvalue=(1-f.cdf(par_fvalue,dfn=1,dfd=forward_sel_df['df_resid']))
forward_sel_df['par_fvalue']=par_fvalue
forward_sel_df['par_f_pvalue']=par_f_pvalue
forward_sel_df.head(10)
```
